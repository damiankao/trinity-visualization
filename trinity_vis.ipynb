## Rendering component graphs from Trinity assembler output

Requires graphviz, which can be downloaded [here](http://www.graphviz.org/Download..php).

Also requires two python packages, [networkx](https://networkx.github.io/) and [pygraphviz](http://pygraphviz.github.io/) which can be installed with pip:

`pip install networkx`

`pip install pygraphviz`.

Use `graphComp()` to graph all nodes of the component. The root node is colored in red.

Use `networkComp()` to graph just the essential nodes of the component that contain alternate branching or used in the probable path prediction. The root node is the '-1' node. The probable paths are highlighted in different colors. The directionality of the edges are specified with a thicker edge line near the outward node.


```python
import os, sys
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline 
    
def graphComp(cid):
    path = 'resource/Component_bins/Cbin0/'
    graphData = [x.split() for x in open(path + 'c' + str(cid) + '.graph.out').read().strip().split('\n')[1:]]
    graph = nx.Graph()

    for item in graphData:
        current = item[0]
        seq = item[-2]
        graph.add_node(current, seq = seq, color='green', size=int(item[2]) * 10)
    graph.add_node('-1',seq = '', color='red', size = 1000)

    for item in graphData:
        graph.add_edge(item[1],item[0])
    plt.figure(figsize=[30,20])
    nx.draw_graphviz(graph, prog='sfdp',
                     node_size = [graph.node[n]['size'] for n in graph.nodes()],
                     node_color = [graph.node[n]['color'] for n in graph.nodes()])
    
def networkComp(cid):
    prepath = 'resource/Component_bins/Cbin0/'
    graphData = [x.split() for x in open(prepath + 'c' + str(cid) + '.graph.out').read().strip().split('\n')[1:]]
    graph = nx.MultiDiGraph()
    
    #iterate through graph once to get all essential nodes
    path = []
    nodes = []
    for item in graphData:
        current = item[0]
        prev = item[1]
        seq = item[-2]
        
        if len(path) == 0:
            path = [prev, current]
        else:
            if path[-1] == prev:
                path.append(current)
            else:
                nodes.append(path[0])
                nodes.append(path[-1])
                path = [prev, current]
    
    nodes.append(path[0])
    nodes.append(path[-1])
    
    #get essential nodes from probable path file
    probPaths = [[node.split(':')[0] for node in line.split('path=')[-1][1:-1].split()]\
                  for line in open(prepath + 'c' + str(cid) + '.graph.allProbPaths.fasta').read().strip().split('\n') \
                  if line[0] == '>']
    nodes.extend(sum(probPaths,[]))
    nodes = list(set(nodes))
    #iterate through again to connect essential
    path = []
    for item in graphData:
        current = item[0]
        prev = item[1]
        seq = item[-2]
        
        if len(path) == 0:
            path = [prev, current]
        else:
            if path[-1] == prev:
                path.append(current)
            else:
                addPath = []
                for node in path:
                    if node in nodes:
                        addPath.append(node)
                graph.add_path(addPath)
                path = [prev, current]
    addPath = []
    for node in path:
        if node in nodes:
            addPath.append(node)
    graph.add_path(addPath)
    
    plt.figure(figsize=[25,10])
    plt.axis('off')
    plt.rcParams['text.usetex'] = False
    pos = nx.graphviz_layout(graph,prog='sfdp')
    
    nx.draw_networkx_nodes(graph,pos,node_size=5000, node_color='white')
    nx.draw_networkx_edges(graph,pos,width=1)
    nx.draw_networkx_labels(graph,pos)
    colors = ['red','blue','green','purple']
    
    print 'Number of probable paths: ', len(probPaths)
    for y, probPath in enumerate(probPaths):
        edgelist = [('-1',probPath[0])]
        for i in range(0,len(probPath) - 1, 1):
            shortest = nx.shortest_path(graph, probPath[i],probPath[i+1])
            [edgelist.append((shortest[a],shortest[a+1])) for a in range(0,len(shortest) - 1, 1)]
        edgelist.append((probPath[-1],graph.neighbors(probPath[-1])[0]))
        nodeList = sum(edgelist,())
        edgePos = []
        for k,v in pos.items():
            if k != '-1' and k in nodeList:
                if graph.node[k].has_key('path'):
                    edgePos.append((k, (v[0], graph.node[k]['path'] + 3)))
                else:
                    edgePos.append((k, (v[0], v[1])))
                    graph.node[k]['path'] = v[1] + 3
            if k == '-1':
                edgePos.append(('-1',(pos['-1'][0],pos['-1'][1])))
        edgePos = dict(edgePos)
        nx.draw_networkx_edges(graph,edgePos,edgelist = edgelist, width = 5, alpha=0.4,edge_color=colors.pop())

networkComp('445')
graphComp('445')
```
